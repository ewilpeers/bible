```python
def split_str_tokened(input_string, tokens):
    result = []
    current_string = input_string
    
    for token in tokens:
        # Atrodam pirmo marķiera parādīšanās vietu
        token_index = current_string.find(token)
        #print(token_index)
        #print(f"::str:: {current_string} \n::token::  {token}")
        # Ja marķieris netiek atrasts, pārtraucam procesu
        if token_index == -1:
            break
            
        # Sadalām virkni: 
        # pirmā daļa līdz marķiera beigām (ieskaitot pašu marķieri)
        # otrā daļa - viss, kas paliek aiz marķiera
        split_pos = token_index + len(token)
        part = current_string[:split_pos]
        current_string = current_string[split_pos:].strip()

        result.append(part)
        

    if current_string:
        result.append(current_string)
    #print(result)
    #print("-"*50)
    #print(tokens)
    return result
```

```python
import unicodedata
import re

# Hebrew letters that commonly get dropped at word endings by AI
HEBREW_FINAL_LETTERS = set("ךםןףץ")  # final kaf, mem, nun, pe, tsadi
HEBREW_SAMEK = "ס"
HEBREW_PE_FINAL = "ף"
HEBREW_SAMEK_STANDALONE = {"פ", "ס"}  # paragraph markers often standalone

def hebrew_char_count(s):
    """Count only base Hebrew letters, ignoring diacritics/cantillation."""
    return sum(1 for c in s if '\u05D0' <= c <= '\u05EA')

def strip_diacritics(s):
    """Remove all Hebrew diacritics, vowels, cantillation — keep base letters only."""
    return ''.join(c for c in s if '\u05D0' <= c <= '\u05EA')

def is_single_diacritic_diff(ref, json_w):
    """True if words differ only by diacritics/cantillation marks."""
    return strip_diacritics(ref) == strip_diacritics(json_w)

def is_one_char_diff(ref, json_w):
    """True if base letter sequences differ by exactly one character."""
    r = strip_diacritics(ref)
    j = strip_diacritics(json_w)
    if abs(len(r) - len(j)) > 1:
        return False
    # same length: exactly one substitution
    if len(r) == len(j):
        return sum(a != b for a, b in zip(r, j)) == 1
    # one longer: one insertion/deletion
    longer, shorter = (r, j) if len(r) > len(j) else (j, r)
    i = 0
    diffs = 0
    for c in longer:
        if i < len(shorter) and shorter[i] == c:
            i += 1
        else:
            diffs += 1
    return diffs == 1

def is_missing_final_letter(ref, json_w):
    """True if json is missing a final Hebrew letter that ref has."""
    r = strip_diacritics(ref)
    j = strip_diacritics(json_w)
    return len(r) == len(j) + 1 and r[:-1] == j and r[-1] in HEBREW_FINAL_LETTERS

def autofix_hebrew_json(key, verse_data, ref_hb_words, verbose=True):
    """
    Reads a verse JSON file, compares hebrew fields against ref_hb_words,
    and auto-fixes:
      1. Diacritics-only differences → replace with ref word
      2. Single base-letter difference → replace with ref word
      3. Missing final letter (samek/pe/finals) → replace with ref word
      4. Missing word (ref has more) → insert minimal JSON entry
    Writes back without reformatting (compact per-word lines preserved).
    """
    verse_num = key[2]

    mapping_hb_words = verse_data.get("hebrew_words", [])

    fixes = []
    ref_list = [str(ref_hb_words.iloc[j]["form"]) for j in range(len(ref_hb_words))]
    json_list = [w.get("hebrew", "") for w in mapping_hb_words]
    PARAGRAPH_MARKERS = {"ס", "פ"}

    # --- Pass 1: fix existing words where counts match ---
    if len(ref_list) == len(json_list):
        for i, (ref_w, json_w) in enumerate(zip(ref_list, json_list)):
            if ref_w == json_w:
                continue
            if (is_single_diacritic_diff(ref_w, json_w) or
                is_one_char_diff(ref_w, json_w) or
                is_missing_final_letter(ref_w, json_w)):
                if verbose:
                    print(f"  ✏️🇮🇱  [{i+1}] '{json_w}' → '{ref_w}'")
                mapping_hb_words[i]["hebrew"] = ref_w
                fixes.append(("replace", i, json_w, ref_w))
                
    # --- Pass 2: ref has exactly one more word than json ---
    elif len(ref_list) == len(json_list) + 1:
        last_ref = strip_diacritics(ref_list[-1])
        if last_ref in PARAGRAPH_MARKERS:
            # missing paragraph marker at the end — just append it
            if verbose:
                print(f"  ➕🇮🇱 inserting missing paragraph marker '{ref_list[-1]}'")
            mapping_hb_words.append({"hebrew": ref_list[-1], "greek": [], "latvian": []})
            fixes.append(("insert_marker", len(ref_list)-1, ref_list[-1]))
        else:
            # single Hebrew word missing somewhere — walk both lists to find where
            inserted = False
            j = 0  # json index
            for ri in range(len(ref_list)):
                if j < len(json_list) and (ref_list[ri] == json_list[j] or
                    is_single_diacritic_diff(ref_list[ri], json_list[j]) or
                    is_one_char_diff(ref_list[ri], json_list[j]) or
                    is_missing_final_letter(ref_list[ri], json_list[j])):
                    # also fix diacritics/small diffs while we're here
                    if ref_list[ri] != json_list[j]:
                        if verbose:
                            print(f"  ✏️🇮🇱  [{j+1}] '{json_list[j]}' → '{ref_list[ri]}'")
                        mapping_hb_words[j]["hebrew"] = ref_list[ri]
                        fixes.append(("replace", j, json_list[j], ref_list[ri]))
                    j += 1
                else:
                    # this ref word has no match in json — insert it
                    if not inserted:
                        new_entry = {"hebrew": ref_list[ri], "greek": [], "latvian": []}
                        mapping_hb_words.insert(j, new_entry)
                        json_list.insert(j, ref_list[ri])  # keep lists in sync
                        inserted = True
                        if verbose:
                            print(f"  ➕🇮🇱 inserting missing word '{ref_list[ri]}' at position {ri}")
                        fixes.append(("insert_word", ri, ref_list[ri]))
                        j += 1
                    else:
                        if verbose:
                            print(f"  ⚠️🇮🇱 multiple mismatches after insert — skipping")
                        return False, fixes, verse_data
    else:
        if verbose:
            print(f"  ⚠️🇮🇱 count mismatch ref={len(ref_list)} vs json={len(json_list)} — skipping")
        return False, fixes, verse_data
        
    # update the verse in place
    verse_data["hebrew_words"] = mapping_hb_words

    return True, fixes, verse_data
```

### fixer for greek

```python
def strip_greek_diacritics(s):
    """Remove Greek diacritics/accents, keep base letters only."""
    return ''.join(
        c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn'  # Mn = nonspacing mark
    )

def strip_greek_deep(s):
    """Strip diacritics, spaces, and modifier symbols for robust comparison."""
    return ''.join(
        c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) not in ('Mn', 'Sk', 'Zs') and c not in ' ΄᾿'
    )

def is_greek_diacritic_diff(ref, json_w):
    """True if Greek words differ only by accents/diacritics.
    Skip single-char words where diacritic changes the word entirely."""
    stripped_ref = strip_greek_diacritics(ref)
    stripped_json = strip_greek_diacritics(json_w)
    # don't fix single-letter words — ἡ/ἢ/ἦ are different words
    if len(stripped_ref) <= 1:
        return False
    return stripped_ref == stripped_json

def autofix_greek_and_leftovers(key, verse_data, ref_strongs, verbose=True):
    """
    1. For each hebrew_word entry, fix Greek words that lost diacritics.
    2. Collect all Greek words from ref_strongs that are not mapped
       in any hebrew_word's greek list, and add missing ones to leftover_greek.
    ref_strongs: DataFrame with 'form' column (Greek LXX words for this verse).
    """
    if not hasattr(ref_strongs, 'iloc'):
        return False  # no strongs data for this verse
        
    verse_num = key[2]

    mapping_hb_words = verse_data.get("hebrew_words", [])
    leftover_greek = verse_data.get("leftover_greek", [])

    fixes = []
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~ define main counter dict
    # Build ref greek word list from strongs (preserving duplicates)
    #APB has space seperated because they provide strongs numbers for multiple words, with non uniform order
    #or matchpairs between words - strongs..
    ref_counts = {}
    for j in range(len(ref_strongs)):
        for w in str(ref_strongs.iloc[j]["form"]).split(' '):
            if not w or w == '-': continue 
            if w not in ref_counts:
                ref_counts[w]=1
            else:
                ref_counts[w]+=1

    #pass 1. direct match, mark as used.        
    untouchables_lamb_mark = []
    for gwi, gw in enumerate(mapping_hb_words):
        lvs = gw.get('greek', [])
        untouchables_lamb_mark.append([False]*len(lvs))
        for li, item in enumerate(lvs):
            if not item or item == '-':
                continue
            if item in ref_counts:
                untouchables_lamb_mark[gwi][li]=True
                if ref_counts[item] == 1:
                    del ref_counts[item]
                else:
                    ref_counts[item] -=1

    #pass 2. mark the soft matches, mark first found pair as used.
    #untouchables_lamb_mark mark which ones will not be removed in pass 3 (remove extra words not present in dataset)
    #now lets match softly the rest of mappings
    # --- Build unmapped ref counts (stripped ignore case) ---
    ref_stripped_counts = {}
    for w, cnt in ref_counts.items():
        sw = strip_greek_deep(w).strip().lower()
        if sw not in ref_stripped_counts:
            ref_stripped_counts[sw] = cnt, [w]*cnt
        else:
            ref_stripped_counts[sw] = ref_stripped_counts[sw][0] + cnt, ref_stripped_counts[sw][1] + [w]*cnt

    for gwi, gw in enumerate(mapping_hb_words):
        lvs = gw.get('greek', [])
        for li, item in enumerate(lvs):
            if(untouchables_lamb_mark[gwi][li]): continue
            if not item or item == '-':
                continue
                
            cmp = strip_greek_deep(item).strip().lower()
            if cmp in ref_stripped_counts:
                untouchables_lamb_mark[gwi][li]=True
                count, forms = ref_stripped_counts[cmp]
                form0 = forms.pop(0) #remove the first form, which is now used, from the list of available forms for this stripped match
                mapping_hb_words[gwi]['greek'][li] = form0 #replace with the form from dataset, which has diacritics and caps
                if (verbose_all or verbose_gk) and not silence_gk_edit:
                    print(f"  ✏️🇬🇷 v{verse_num} w{gwi+1}: '{item}' → '{form0}'")
                fixes.append(('gk_mapped_edit', key, item))
                if count == 1:
                    del ref_stripped_counts[cmp]
                else:
                    ref_stripped_counts[cmp] = count - 1, forms
                #exception here ensures data integrity.
                if ref_counts[form0] == 1:
                    del ref_counts[form0]
                else:
                    ref_counts[form0] -=1
                    
# pass 3 EXODUS 12 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    for gwi, gw in enumerate(mapping_hb_words):
        lvs = gw.get('greek', [])
        new_lvs = []
        for li, item in enumerate(lvs):
            if(untouchables_lamb_mark[gwi][li]): new_lvs.append(item)
            else:
                if verbose_all or verbose_gk:
                    print(f"  ➖🇬🇷 v{verse_num} w{gwi+1}: removed extra '{item}'")
                fixes.append(('gk_mapped_strip', key, item))
        mapping_hb_words[gwi]['greek'] = new_lvs

    # pass 4.A then go with the leftovers.
    untouchables_LO = [False]*len(leftover_greek) 
    for li, item in enumerate(leftover_greek):
        ##breakpoint()
        if not item or item == '-':
            continue
        if item in ref_counts:
            untouchables_LO[li]=True
            stripped_item = strip_greek_deep(item).strip().lower()
            if ref_counts[item] == 1:
                del ref_counts[item]
                number, items = ref_stripped_counts[stripped_item]
                if item in items:
                    items.remove(item)
                    if number == 1:
                        del ref_stripped_counts[stripped_item]
                    else:
                        ref_stripped_counts[stripped_item] = number-1, items
            else:
                ref_counts[item] -=1
                number, items = ref_stripped_counts[stripped_item]
                #match first item. 
                items.remove(item)
                ref_stripped_counts[stripped_item] = number-1, items
                if ref_stripped_counts[stripped_item][0] == 0:
                    del ref_stripped_counts[stripped_item]
        else:
            cmp = strip_greek_deep(item).strip().lower()
            if cmp in ref_stripped_counts:
                untouchables_LO[li]=True
                count, forms = ref_stripped_counts[cmp]
                form0 = forms.pop(0)
                #HERE REF_stripped is used also as initial array, so it can be they match already
                if(form0!=item):
                    leftover_greek[li] = form0 #replace with the form from dataset, which has diacritics and caps
                    if (verbose_all or verbose_gk) and not silence_gk_edit:
                        print(f"  ✏️🇬🇷 v{verse_num} leftover #{li+1}: '{item}' → '{form0}'")
                    fixes.append(('gk_leftover_edit', key, [item]))
                if count == 1:
                    del ref_stripped_counts[cmp]
                    del ref_counts[form0]
                else:
                    ref_stripped_counts[cmp] = count - 1, forms
                    ref_counts[form0] -=1

    # pass 4.B EXODUS 12 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    new_leftovers = []
    for li, item in enumerate(leftover_greek):
        if(untouchables_LO[li]): new_leftovers.append(item)
        else:
            if verbose_all or verbose_gk:
                print(f"  ➖🇬🇷 v{verse_num} leftover #{li+1}: removed extra '{item}'")
            fixes.append(('gk_leftover_strip', key, item))
    leftover_greek[:] = new_leftovers

    # pass 5. then add the missing ones from ref_counts to leftovers, no matter if ne or not, because if they were in mappings, they would be matched already, and if they were in leftovers, they would be matched in pass 4.
    for w, count in ref_counts.items():
        for _ in range(count):
            if verbose_all or verbose_gk:
                print(f"  ➕🇬🇷 v{verse_num} missing leftover: '{w}'")
            leftover_greek.append(w)
            fixes.append(('gk_leftover_add', key, [w]))

    # update and write back
    verse_data["hebrew_words"] = mapping_hb_words
    verse_data["leftover_greek"] = leftover_greek

    return fixes, verse_data
```

```python
import unicodedata, re
from collections import Counter, defaultdict

LV_CHARS = "A-Za-zāēīūģķļņšžčĀĒĪŪĢĶĻŅŠŽČ"
LV_REGEX = re.compile(f"[{LV_CHARS}]+")

def _lv_tokens(text: str):
    """Return list of Latvian word-tokens from any string."""
    return LV_REGEX.findall(unicodedata.normalize("NFC", text))

def _strip_lv_diacritics(s: str) -> str:
    """
    Lower-case and strip Latvian diacritics for comparison.
    e.g. 'Ābraham' → 'abraham', 'šķēp' → 'skep'
    """
    return "".join(
        c for c in unicodedata.normalize("NFD", s.lower())
        if unicodedata.category(c) != "Mn"
    )
def strip_latvian_deep(s: str) -> str:
    return _strip_lv_diacritics(s)
def autofix_latvian_and_leftovers(key, verse_data, ref_lv_words_raw, verbose=True):
    """
    Pass 1 – restore diacritics (ā ē ī ū š ž č ģ ķ ļ ņ) in
             hebrew_words[*]['latvian'] using l65_df canonical forms.
             Only fixes single-token entries where the match is unambiguous.
    Pass 2 – append any ref tokens not yet accounted for to leftover_latvian.
    Writes back in the same compact format as the Hebrew / Greek fixers.
    ref_lv_words_raw : list[str]  — tokens from _lv_tokens(l65_text)
    """
    if not ref_lv_words_raw:
        print(f"{key} !!!!!! NO LV REF WORDS RAW!!!!!!")
        return [], verse_data
    verse_num = key[2]
    mapping_hb_words = verse_data.get("hebrew_words", [])
    leftover_latvian = verse_data.get("leftover_latvian", [])
    #has_ne_prefix - whether JSON has at least i_one "ne" or "ne-"
    #ne_ref_count - how many words in ref (dataset) i_starts with "ne"
    #ne_json_count - how many words in JSON i_starts with "ne" (mappings + lovers)
    fixes = []


    has_ne_prefix=False
    for w in leftover_latvian:
        if len(w) <4:
            if w.lower in ('ne', 'ne-'):
                has_ne_prefix=True
                break
    # Build ref latvian word list from strongs (preserving duplicates)
    #APB has space seperated because they provide strongs numbers for multiple words, with non uniform order
    #or matchpairs between words - strongs..
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~ define main counter dict
    ref_counts = {}
    #sadly to avoid cyclyc ref, the count will be here always, always slicing.
    #now no trim (strip)  here, assuming ds words are no whitespaces pad
    ne_ref_count=0
    for i, ww in enumerate(ref_lv_words_raw):
        for w in ww.split(' '):
    #there is particle "ne" in Latvian too: ne vienu ne otru..
    #nesa nes nesen, will cause extra ne to preserve..
            #ne_ref_count += sum(1 if w.lower().startswith("ne") else 0 for w in ww)
            ne_ref_count += 1 if w.lower().startswith("ne") else 0
            if not w or w == '-': continue 
            if w not in ref_counts:
                ref_counts[w]=1
            else:
                ref_counts[w]+=1

    #pass 1. direct match, mark as used.  
    #so that they are not more, but single ne per multiple roots is allowed
    ne_json_count = 0      
    untouchables_lamb_mark = []
    for gwi, gw in enumerate(mapping_hb_words):
        lvs = gw.get('latvian', [])
        untouchables_lamb_mark.append([False]*len(lvs))
        for li, item in enumerate(lvs):
            if not item or item == '-':
                continue
            if item in ref_counts:
                untouchables_lamb_mark[gwi][li]=True
                if item[:2].lower()=='ne':
                    ne_json_count+=1
                if ref_counts[item] == 1:
                    del ref_counts[item]
                else:
                    ref_counts[item] -=1
            if not has_ne_prefix and len(item)<4:
                    #nes
                if (item[0]=='n' or item[0]=='N') and len(item)>1 and \
                     (item[1]=='e' or item[1]=='E')\
                    and item.lower() in ('ne-', 'ne'):
                    has_ne_prefix = True

#~~~~~~~~pass 2. mark the soft matches, mark first found pair as used.
    #untouchables_lamb_mark mark which ones will not be removed in pass 3 (remove extra words not present in dataset)
    #now lets match softly the rest of mappings
    # --- Build unmapped ref counts (stripped ignore case) ---
    ref_stripped_counts = {}
    
    for w, cnt in ref_counts.items():
        sw = strip_latvian_deep(w).strip().lower()
        if sw not in ref_stripped_counts:
            ref_stripped_counts[sw] = cnt, [w]*cnt
        else:
            ref_stripped_counts[sw] = ref_stripped_counts[sw][0] + cnt, ref_stripped_counts[sw][1] + [w]*cnt

    for gwi, gw in enumerate(mapping_hb_words):
        lvs = gw.get('latvian', [])
        for li, item in enumerate(lvs):
            if(untouchables_lamb_mark[gwi][li]): continue
            if not item or item == '-':
                continue
            cmp = strip_latvian_deep(item).strip().lower()
            if(len(cmp)==2 and len(item.strip())==3):
                sitem = item.strip()
                if (sitem[0]=='n' or sitem[0]=='N')  and (sitem[1] in ('e', 'E')) and sitem[2] =='-':
                    #THIS IS MARKER FOR NE-PREFIX, should not match to the ne particle!
                    cmp = cmp+"-"
                    
            if cmp in ref_stripped_counts:
                untouchables_lamb_mark[gwi][li]=True
                count, forms = ref_stripped_counts[cmp]
                form0 = forms.pop(0) #remove the first form, which is now used, from the list of available forms for this stripped match
                mapping_hb_words[gwi]['latvian'][li] = form0 #replace with the form from dataset, which has diacritics and caps
                if item != form0:
                    if (verbose_all or verbose_lv) and not silence_lv_edit:
                        print(f"  ✏️🇱🇻 v{verse_num} w{gwi+1}: '{item}' → '{form0}'")
                    fixes.append(('lv_mapped_edit', key, item))
                if form0[:2].lower()=='ne':
                    ne_json_count+=1
                if count == 1:
                    del ref_stripped_counts[cmp]
                else:
                    ref_stripped_counts[cmp] = count - 1, forms
                untouchables_lamb_mark[gwi][li]=True
                #exception here ensures data integrity.
                if ref_counts[form0] == 1:
                    del ref_counts[form0]
                else:
                    ref_counts[form0] -=1
            elif has_ne_prefix and (item[0]=='n' or item[0]=='N')  \
                and len(item)>1 and (item[1] in ('e', 'E')) and \
                    (True if len(item)==2 else item[2]=='-'):
                 #allow the ne prefix or ne particle to be there, unless they are more than matching ne prefixes/particles at dataset! then those extra must be removed
                 # now the case is lost, but that is the sacrifice required..
                ne_json_count +=1
                untouchables_lamb_mark[gwi][li]=True
            elif has_ne_prefix:
                # ne prefix match mode. single ne means multiple stems are allowed to match,
                #  but each dataset word can only be matched once, so we need to check the count 
                # and forms list for each match.
                refcopy = ref_stripped_counts.copy()
                #can raise RuntimeError: dictionary changed size during iteration if we try to modify the dict while iterating, so we need to iterate over a copy of the dict keys.
                for cmp in refcopy:
                    if cmp.startswith('ne') and len(cmp)>2 and cmp[2:]==strip_latvian_deep(item).strip().lower():
                        untouchables_lamb_mark[gwi][li]=True
                        count, forms = ref_stripped_counts[cmp]
                        form0 = forms.pop(0)
                        if(form0[2:]!=item):
                            mapping_hb_words[gwi]['latvian'][li] = form0[2:] #replace with the form from dataset, which has diacritics and caps
                            if (verbose_all or verbose_lv) and not silence_lv_edit:
                                print(f"  ✏️🇱🇻 v{verse_num} w{gwi+1}: '{item}' → '{form0}' (ne-compound match)")
                            fixes.append(('lv_mapped_edit', key, item))
                        if count == 1:
                            del ref_stripped_counts[cmp]
                        else:
                             ref_stripped_counts[cmp] = count - 1, forms
                        #exception here ensures data integrity.
                        if ref_counts[form0] == 1:
                            del ref_counts[form0]
                        else:
                            ref_counts[form0] -=1
                        break

                    
# pass 3 EXODUS 12 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    for gwi, gw in enumerate(mapping_hb_words):
        lvs = gw.get('latvian', [])
        new_lvs = []
        for li, item in enumerate(lvs):
            if(untouchables_lamb_mark[gwi][li]): new_lvs.append(item)
            else:
                if verbose_all or verbose_lv:
                    print(f"  ➖🇱🇻 v{verse_num} w{gwi+1}: removed extra '{item}'")
                fixes.append(('lv_mapped_strip', key, item))
        mapping_hb_words[gwi]['latvian'] = new_lvs

    # pass 4.A then go with the leftovers.
    untouchables_LO = [False]*len(leftover_latvian) 
    for li, item in enumerate(leftover_latvian):
        if not item or item == '-':
            continue
        if item in ref_counts:
            untouchables_LO[li]=True
            if item[:2].lower()=='ne':
                ne_json_count+=1
            stripped_item = strip_latvian_deep(item).strip().lower()
            if ref_counts[item] == 1:
                del ref_counts[item]
                number, items = ref_stripped_counts[stripped_item]
                if item in items:
                    items.remove(item)
                    if number == 1:
                        del ref_stripped_counts[stripped_item]
                    else:
                        ref_stripped_counts[stripped_item] = number-1, items
            else:
                ref_counts[item] -=1
                number, items = ref_stripped_counts[stripped_item]
                #match first item.
                items.remove(item)
                ref_stripped_counts[stripped_item] = number-1, items
                if ref_stripped_counts[stripped_item][0] == 0:
                    del ref_stripped_counts[stripped_item]
        else:
            cmp = strip_latvian_deep(item).strip().lower()
            if cmp in ref_stripped_counts:
                untouchables_LO[li]=True
                count, forms = ref_stripped_counts[cmp]
                form0 = forms.pop(0)
                if form0[:2].lower()=='ne':
                    ne_json_count+=1
                #HERE REF_stripped is used also as initial array, so it can be they match already
                if(form0!=item):
                    leftover_latvian[li] = form0 #replace with the form from dataset, which has diacritics and caps
                    if (verbose_all or verbose_lv) and not silence_lv_edit:
                        print(f"  ✏️🇱🇻 v{verse_num} leftover #{li+1}: '{item}' → '{form0}'")
                    fixes.append(('lv_leftover_edit', key, [item]))
                if count == 1:
                    del ref_stripped_counts[cmp]
                    del ref_counts[form0]
                else:
                    ref_stripped_counts[cmp] = count - 1, forms
                    ref_counts[form0] -=1
            elif has_ne_prefix and (item[0]=='n' or item[0]=='N')  \
                and len(item)>1 and (item[1] in ('e', 'E')) and \
                    (True if len(item)==2 else item[2]=='-'):
                #allow the ne prefix or ne particle to be there, unless they are more than matching ne prefixes/particles at dataset! then those extra must be removed
                # now the case is lost, but that is the sacrifice required..
                ne_json_count +=1
                untouchables_LO[li]=True
            elif has_ne_prefix:
                refcopy = ref_stripped_counts.copy()
                for cmp in refcopy:
                    if cmp.startswith('ne') and len(cmp)>2 and cmp[2:]==strip_latvian_deep(item).strip().lower():
                        untouchables_LO[li]=True
                        count, forms = ref_stripped_counts[cmp]
                        form0 = forms.pop(0)
                        if(form0[2:]!=item):
                            leftover_latvian[li] = form0[2:] #replace with the form from dataset, which has diacritics and caps
                            if (verbose_all or verbose_lv) and not silence_lv_edit:
                                #if(li==5):
                                #    print([hex(ord(c)) for c in item])
                                #    print([hex(ord(c)) for c in form0])
                                print(f"  ✏️🇱🇻 v{verse_num} leftover #{li+1}: '{item}' → '{form0}' (ne-compound match)")
                            fixes.append(('lv_leftover_edit', key, [item]))
                        if count == 1:
                            del ref_stripped_counts[cmp]
                            del ref_counts[form0]
                        else:
                            ref_stripped_counts[cmp] = count - 1, forms
                            ref_counts[form0] -=1             
                        break
    # pass 4.B EXODUS 12 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    new_leftovers = []
    for li, item in enumerate(leftover_latvian):
        if(untouchables_LO[li]): new_leftovers.append(item)
        else:
            if verbose_all or verbose_lv:
                print(f"  ➖🇱🇻 v{verse_num} leftover #{li+1}: removed extra '{item}'")
            fixes.append(('lv_leftover_strip', key, item))
    leftover_latvian[:] = new_leftovers

    # pass 4444444.5. then add the missing ones from ref_counts to leftovers, no matter if ne or not, because if they were in mappings, they would be matched already, and if they were in leftovers, they would be matched in pass 4.
    for w, count in ref_counts.items():
        if(w.startswith('ne')): ne_json_count += count
        for _ in range(count):
            if verbose_all or verbose_lv:
                print(f"  ➕🇱🇻 v{verse_num} missing leftover: '{w}'")
            leftover_latvian.append(w)
            fixes.append(('lv_leftover_add', key, [w]))
            
            
    # pass 4.C Mark with lamb bld -> "ne"/"ne-" !!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    # so that removal is from the end first, from lovers, then mappings
    hadDiff = False
    diffne = ne_json_count - ne_ref_count
    if ne_json_count>ne_ref_count:
        print(f'{key} JS {ne_json_count} {ne_ref_count} DF')
        hadDiff = True
        untouchables_LO = [False]*len(leftover_latvian)
        #remove excess ne- from mappings first, then from leftovers if still excess
        #for li, item in reversed(list(enumerate(leftover_latvian))):
        for li in range(len(leftover_latvian) - 1, -1, -1):
            item = leftover_latvian[li]
            if (item[0]=='n' or item[0]=='N') and len(item)>1 and (item[1] in ('e', 'E')) and \
                    (True if len(item)==2 else item[2]=='-'):
                if diffne>0:
                    untouchables_LO[li]=False
                    diffne-=1
                    ne_json_count-=1
                else:
                    untouchables_LO[li]=True
            else:
                untouchables_LO[li]=True
        if(diffne>0):
            untouchables_lamb_mark = []
            #for gwi, gw in reversed(list(enumerate(mapping_hb_words))):
            for gwi in range(len(mapping_hb_words) - 1, -1, -1):
                gw = mapping_hb_words[gwi]
                lvs = gw.get('latvian', [])
                untouchables_lamb_mark.insert(0, [False]*len(lvs))
                for li, item in enumerate(lvs):
                    if (item[0]=='n' or item[0]=='N') and len(item)>1 and (item[1] in ('e', 'E')) and \
                            (True if len(item)==2 else item[2]=='-'):
                        if diffne>0:
                            untouchables_lamb_mark[0][li]=False
                            diffne-=1
                            ne_json_count-=1
                        else:
                            untouchables_lamb_mark[0][li]=True
                    else:
                        untouchables_lamb_mark[0][li]=True


    # pass 4.D EXODUS 12 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    if hadDiff:
        for gwi, gw in enumerate(mapping_hb_words):
            lvs = gw.get('latvian', [])
            new_lvs = []
            for li, item in enumerate(lvs):
                if(untouchables_lamb_mark[gwi][li]): new_lvs.append(item)
                else:
                    if verbose_all or verbose_lv:
                        print(f"  ➖🇱🇻 v{verse_num} w{gwi+1}: removed extra ne '{item}'")
                    fixes.append(('lv_mapped_ne_strip', key, item))
            mapping_hb_words[gwi]['latvian'] = new_lvs
            new_leftovers = []
            
        for li, item in enumerate(leftover_latvian):
            if(untouchables_LO[li]): new_leftovers.append(item)
            else:
                if verbose_all or verbose_lv:
                    print(f"  ➖🇱🇻 v{verse_num} leftover #{li+1}: removed extra ne '{item}'")
                fixes.append(('lv_leftover_ne_strip', key, item))
        leftover_latvian[:] = new_leftovers
        


    # STORAGE> update and return
    verse_data["hebrew_words"] = mapping_hb_words
    verse_data["leftover_latvian"] = leftover_latvian

    return fixes, verse_data
```

### diff renderer definition

```python
GREEN = '\033[92m'
RED_W =  '\x1b[38;5;88m\x1b[48;5;7m'
DIFF_BLUE =  '\x1b[38;5;12m\x1b[48;15;7m'
DIFF_RED =  '\x1b[38;5;196m\x1b[48;15;7m'
RESET = '\033[0m'
def diff_hebrew_vs_json(ref_hb_words, mapping_hb_words, key):
    aminlen = min(len(ref_hb_words), len(mapping_hb_words))
    if(len(ref_hb_words)!= len(mapping_hb_words)):
        print(f"❌🇮🇱 {RED_W}{key[2]} Hebrew word count mismatch at {key}:{RESET} Ref {len(ref_hb_words)} vs JSON {len(mapping_hb_words)}")


    hadDiff = False

    for j in range(aminlen):
        ref_str  = str(ref_hb_words.iloc[j]["form"])
        json_str = mapping_hb_words[j].get("hebrew", "")
        inum= int(mapping_hb_words[j].get("index", str(j+1)))

        if ref_str == json_str:
            # identical — print plain
            if hadDiff:
                print(f"  {j+1:<4} {GREEN}{ref_str}{RESET}")
        else:
            if not hadDiff:
                print(f"Δ❌Δ🇮🇱 v{RED_W}{key[2]}{RESET} wo {DIFF_RED}{j+1}{RESET} rd : {key}  Hebrew  word mismatch")
                print(f"  {'#':<4} {'ref':<30} {'json'}")
            hadDiff = True
            # diff char by char
            minlen = min(len(ref_str), len(json_str))
            s_ref = s_json = ""
            for i in range(minlen):
                if ref_str[i] == json_str[i]:
                    s_ref  += f"{GREEN}{ref_str[i]}{RESET}"
                    s_json += f"{GREEN}{json_str[i]}{RESET}"
                else:
                    s_ref  += f"{DIFF_RED}{ref_str[i]}{RESET}"
                    s_json += f"{DIFF_RED}{json_str[i]}{RESET}"
            if len(ref_str) > minlen:
                s_ref  += f"{DIFF_BLUE}{ref_str[minlen:]}{RESET}"
            if len(json_str) > minlen:
                s_json += f"{DIFF_BLUE}{json_str[minlen:]}{RESET}"
            # one word per line to avoid bidi chaos
            print(f"  {inum:<4} ref : {s_ref}")
            print(f"  {'':4} json: {s_json}")

    # surplus words
    if len(ref_hb_words) > aminlen:
        for j in range(aminlen, len(ref_hb_words)):
            form = str(ref_hb_words.iloc[j]["form"])
            print(f"  {j+1:<4} {DIFF_BLUE}ref only : {form}{RESET}")
    if len(mapping_hb_words) > aminlen:
        for j in range(aminlen, len(mapping_hb_words)):
            heb = mapping_hb_words[j].get("hebrew", "")
            print(f"  {j+1:<4} {DIFF_BLUE}json only: {heb}{RESET}")

def diff_greek_vs_json(ref_strongs_df, mapping_hb_words, leftover_greek, key):
    """Compare ref Greek words against JSON's greek arrays + leftover_greek. Report mismatches."""
    if ref_strongs_df is None:
        return
    pre_ref_greek_words = [str(ref_strongs_df.iloc[j]['form']) for j in range(len(ref_strongs_df))]
    ref_greek_words=[]
    for i, w in enumerate(pre_ref_greek_words):
        if not w[:2]=='΄ ':
            if ' ' in w:
                for ww in w.split(' '):
                    ref_greek_words.append(ww)
            else:
                ref_greek_words.append(w)
        else:
            ref_greek_words.append(w)
    
    
    json_greek = []
    for w in mapping_hb_words:
        json_greek.extend(w.get('greek', []))
    json_greek.extend(leftover_greek)
    
    ref_counts = Counter(ref_greek_words)
    json_counts = Counter(json_greek)
    ref_stripped_counts = Counter(strip_greek_diacritics(w) for w in ref_greek_words)
    json_stripped_counts = Counter(strip_greek_diacritics(w) for w in json_greek)
    
    missing = []
    for word, count in ref_counts.items():
        sw = strip_greek_diacritics(word)
        stripped_deficit = ref_stripped_counts[sw] - json_stripped_counts.get(sw, 0)
        if stripped_deficit > 0:
            deficit = count - json_counts.get(word, 0)
            if deficit > 0:
                missing.extend([word] * min(deficit, stripped_deficit))
    
    extra = []
    for word, count in json_counts.items():
        sw = strip_greek_diacritics(word)
        stripped_surplus = json_stripped_counts[sw] - ref_stripped_counts.get(sw, 0)
        if stripped_surplus > 0:
            surplus = count - ref_counts.get(word, 0)
            if surplus > 0:
                extra.extend([word] * min(surplus, stripped_surplus))
    
    if missing:
        if verbose:
            print(f"  ⚠️🇬🇷 v{key[2]} Greek missing in JSON: {missing}")
    if extra:
        if verbose:
            print(f"  ⚠️🇬🇷 v{key[2]} Greek extra in JSON (not in ref): {extra}")
```

## validate lv usagee

```python
import pandas as pd
import time
from pathlib import Path
import json
import os
import json
import re
from collections import Counter
import unicodedata
import subprocess
import sys
# whether to write each of the verses as seperate, splitted from the chapter json, file .txt
# and run renumber_output.py call after each of them
NO_SPLIT_AND_RENUMBER_RUN=1
LOG_RENUMBER_OUTPUT=0

def validate_latvian_usage(verse_text, mappings, leftover_latvian_from_verse_data):
    """
    Pārbauda, vai visi vārdi no latviešu teksta ir atrodami kartējumos.
    Izmanto specifisku latviešu burtu kopu, lai novērstu vārdu sašķelšanos.
    """
    # 1. Normalizējam tekstu uz NFC (lai 'ī' būtu viens simbols, nevis i + macron)
    ####print(verse_text)
    verse_text = unicodedata.normalize('NFC', verse_text)
    
    # Definējam latviešu burtu kopu (lielie un mazie)
    lv_chars = "A-Za-zāēīūģķļņšžčĀĒĪŪĢĶĻŅŠŽČ"
    lv_regex = f"[{lv_chars}]+"

    # 2. Tokenizējam tekstu
    text_words = re.findall(lv_regex, verse_text)
    
    # 3. Savācam visus vārdus no kartējumiem
    mapped_words = []
    
    for m in mappings:
        # Pārbaudām galvenos kartējumus
        if m.get('latvian'):
            for item in m['latvian']:
                if item and item != "-":
                    #TODO: ja ir neatzīst un atzīst orig tulkojumos vienlaikus...
                    # Arī šeit normalizējam un meklējam ar lv_regex
                    clean_item = unicodedata.normalize('NFC', item)
                    #print(re.findall(lv_regex, clean_item))
                    mapped_words.extend(re.findall(lv_regex, clean_item))
        
        if leftover_latvian_from_verse_data:
            for item in leftover_latvian_from_verse_data:
                clean_item = unicodedata.normalize('NFC', item)
                mapped_words.extend(re.findall(lv_regex, clean_item))
    
    text_counts = Counter(text_words)
    mapped_counts = Counter(mapped_words)

    missing = []
    for word, count in text_counts.items():
        if mapped_counts[word] < count:
            lw = word.lower()
            
            # Loģika priedēklim "ne-" (piemēram, 'atzīst' vs 'neatzīst')
            if len(lw) > 2 and lw.startswith("ne"):
                stem = word[2:]  # Saglabājam oriģinālo reģistru saknei
                if mapped_counts[stem] >= count:
                    continue
            
            # Pievienojam trūkstošos gadījumus
            for _ in range(count - mapped_counts[word]):
                missing.append(word)
            
    return missing
```

```python
SHOW_LATVIANS=0
def fix_chapter_json(book, chapter_num, hb_g, strongs_g, lv_g):
    fixes = []
    # Path to the source chapter file: bible/genesis/1.json
    source_path = Path("bible") / book.lower() / f"{chapter_num}.json"
    
    if not source_path.exists():
        print(f"❓ Source {source_path} not found. Skipping.")
        return []

    with open(source_path, "r", encoding="utf-8") as f:
        verses_list = json.load(f)

    expected_verses = sorted(
    v for (b, c, v) in hb_g.groups.keys()
    if b == book and c == int(chapter_num)
    )
    if len(verses_list) != len(expected_verses):
        print(
            f"❌🔥🔥🔥❌ {source_path}: \nJSON has {len(verses_list)} verse(s) \n"
            f"but hb_g expects {len(expected_verses)} \n"
            f"(verses {expected_verses[0]}–{expected_verses[-1]}). \n"
            f"Missing: {expected_verses[len(verses_list):]}\n"
        )
        return []
    #------noooooooooo--writeable copy so that verses_list write attempt does not throw--------
#     The copy would only matter if:
# You were doing verses_list.append(...) or del verses_list[j] mid-loop, or
# You needed the pre-mutation state of sibling verses for comparison.
    #data = verses_list[:]
    hadAnyFixes = False
    for itsverseindex, verse_data in enumerate(verses_list):
        verse_number = int(itsverseindex)+1
        key = (book, int(chapter_num), verse_number)

        # 1. Get Reference Data
        if key not in hb_g.groups:
            print(f"⚠️🇮🇱 Missing Hebrew reference data for {key}.")
            raise Exception(f"HBRW missing - no such chapter {key}")
        else:
            ref_hb = hb_g.get_group(key).sort_values("word")

        ref_strongs_from_map_wrapper = get_greek_text_for_hb_key(key, strongs_g, abp_strongs_g)
        ref_strongs_df = get_greek_df_for_hb_key(key, strongs_g, abp_strongs_g)
        if not ref_strongs_from_map_wrapper:
            if verbose or verbose_greek_missings:
                print(f"⚠️🇬🇷 Missing STRONGS reference data for {key}.")
            ref_strongs=[]
        else:
            ref_strongs = ref_strongs_from_map_wrapper

        # 2. hebrew validation
        mapping_hb_words = verse_data.get('hebrew_words', [])
        #print(f"{key} {len(ref_hb)}, {len(mapping_hb_words)}")
        OK, fixes_hb, verses_list[itsverseindex] = autofix_hebrew_json(key, verses_list[itsverseindex], ref_hb, verbose=verbose)
        if not OK:
            continue
        else:
            hadAnyFixes = hadAnyFixes or len(fixes_hb)>0
            fixes.extend(fixes_hb)
  
        fixes_gr, verses_list[itsverseindex] = autofix_greek_and_leftovers(key, verses_list[itsverseindex], ref_strongs_df, verbose=verbose)
        hadAnyFixes = hadAnyFixes or len(fixes_gr)>0
        fixes.extend(fixes_gr)

        
        mapping_hb_words = verses_list[itsverseindex].get('hebrew_words', [])

        diff_hebrew_vs_json(ref_hb, mapping_hb_words, key)
        diff_greek_vs_json(ref_strongs_df, mapping_hb_words, verse_data.get('leftover_greek', []), key)
            #print(mapping_hb_words)
            #continue
        
        l65_text = get_l65_text_for_hb_key(key, l65_g)
        
        leftover_latvian_from_verse_data = verse_data.get('leftover_latvian', [])
        # 3a. Validation: Latvian Usage (debug only)
        if SHOW_LATVIANS:
            latvian_missing = validate_latvian_usage(l65_text, mapping_hb_words, leftover_latvian_from_verse_data)
            if latvian_missing:
                print(f"⚠️🇱🇻 Latvian mapping incomplete at {key}: Missing {', '.join(latvian_missing)}")

        # 3b. Autofix: restore diacritics + write missing words to leftover_latvian
        ref_lv_tokens = _lv_tokens(l65_text)
        fixes_lv, verses_list[itsverseindex] = autofix_latvian_and_leftovers(
            key, verses_list[itsverseindex], ref_lv_tokens, verbose=verbose
        )
        hadAnyFixes = hadAnyFixes or len(fixes_lv)>0
        fixes.extend(fixes_lv)

    # write back to FILE the changes..        
    if hadAnyFixes:
        def word_to_compact(w):
            heb     = w.get("hebrew",  "")
            greek   = w.get("greek",   [])
            latvian = w.get("latvian", [])
            hb_str = json.dumps(heb,   ensure_ascii=False)
            gr_str = json.dumps(greek,   ensure_ascii=False)
            lv_str = json.dumps(latvian, ensure_ascii=False)
            return f'{{ "hebrew": {hb_str}, "greek": {gr_str}, "latvian": {lv_str} }}'
    
        verse_blocks = []
        for verse in verses_list:
            words         = verse.get("hebrew_words", [])
            compact_words = ",\n      ".join(word_to_compact(w) for w in words)
            lo_lv = json.dumps(verse.get("leftover_latvian", []), ensure_ascii=False)
            lo_gr = json.dumps(verse.get("leftover_greek",   []), ensure_ascii=False)
            verse_blocks.append(
                f'  {{\n    "hebrew_words": [\n      {compact_words}\n    ],\n'
                f'    "leftover_greek": {lo_gr},\n'
                f'    "leftover_latvian": {lo_lv}\n  }}'
            )
    
        wrdata =  "[\n" + ",\n".join(verse_blocks) + "\n]"
        with open(source_path, "w", encoding="utf-8") as f:
            f.write(wrdata)
        if verbose:
            print(f"  ✅🇱🇻 Successfully corrected {source_path}\n{len(fixes)} fix(es) written.")
            
    return fixes
```
