# 1.5 /3 Hebrew (MT) to LXX verse mapping
## Based on SBL Handbook of Style Appendix B and known LXX differences

The LXX (Septuagint) for 1 Samuel (= 1 Kingdoms / Βασιλειων Αʹ) follows English/modern
versification rather than Hebrew MT numbering. This means for the chapter/verse offsets
that exist in the Hebrew MT, the LXX uses the English numbering convention.

Additionally, the LXX of 1 Samuel has significant **content** differences:
- Chapters 17-18: LXX is much shorter, missing ~39 of 88 MT verses
- 1 Sam 13:1 is absent in LXX Sinaiticus
- Various additions/omissions throughout

For books beyond 1 Samuel the same SBL table is used.

The mapping `hb_to_lxx` converts Hebrew MT (book, chapter, verse) keys
to the corresponding LXX (book, chapter, verse) key so the Greek text
can be looked up correctly when the Hebrew numbering differs.

````python
# ============================================================
# Hebrew MT -> LXX verse mapping
# ============================================================
# The LXX follows English versification for chapter/verse offsets.
# Where Hebrew MT has a different chapter/verse boundary, we map
# the Hebrew key to the LXX (=English) key so Greek text lookup works.
#
# Format: hb_to_lxx[heb_key] = lxx_key
# Where no mapping exists, Hebrew key == LXX key (identity)
#
# Source: SBL Handbook of Style, Appendix B (2014), pp. 265-268
# https://matthewbarron.org/bible-versification-compared/
# ============================================================

hb_to_lxx = {}
# Also track LXX verses that have NO Hebrew equivalent (LXX-only content)
lxx_only_verses = set()
# And Hebrew verses with no LXX equivalent
hb_no_lxx = set()

# ----------------------------------------------------------
# 1 SAMUEL  (Hebrew שמואל א = LXX Βασιλειων Αʹ / 1 Kingdoms)
# ----------------------------------------------------------
# English/LXX 20:42 last clause = Hebrew 21:1
# English/LXX 21:1-15 = Hebrew 21:2-16
# So: Hebrew 21:1 -> LXX 20:42 (merged/appended to end of 20:42)
#     Hebrew 21:2 -> LXX 21:1 ... Hebrew 21:16 -> LXX 21:15

# Hebrew 21:1 maps to the tail of LXX 20:42
hb_to_lxx[('1_samuel', 21, 1)] = ('1_samuel', 20, 42)  # merged into end of 20:42
for i in range(2, 16+1):
    hb_to_lxx[('1_samuel', 21, i)] = ('1_samuel', 21, i-1)

# English/LXX 23:29 = Hebrew 24:1
# English/LXX 24:1-22 = Hebrew 24:2-23
hb_to_lxx[('1_samuel', 24, 1)] = ('1_samuel', 23, 29)
for i in range(2, 23+1):
    hb_to_lxx[('1_samuel', 24, i)] = ('1_samuel', 24, i-1)

# LXX 1 Samuel 17-18 missing verses (LXX lacks these MT verses entirely)
# LXX ch17 has: 1-11, 32-40, 42-49, 51-54  (missing 12-31, 41, 50, 55)
# LXX ch18 has: 6-9, 12-16, 20-29 (missing 1-5, 10-11, 17-19, 30)
# These are based on Rahlfs' edition of the LXX (Codex Vaticanus tradition)
for v in range(12, 31+1):
    hb_no_lxx.add(('1_samuel', 17, v))
hb_no_lxx.add(('1_samuel', 17, 41))
hb_no_lxx.add(('1_samuel', 17, 50))
hb_no_lxx.add(('1_samuel', 17, 55))
hb_no_lxx.add(('1_samuel', 17, 56))
hb_no_lxx.add(('1_samuel', 17, 57))
hb_no_lxx.add(('1_samuel', 17, 58))
for v in range(1, 5+1):
    hb_no_lxx.add(('1_samuel', 18, v))
hb_no_lxx.add(('1_samuel', 18, 10))
hb_no_lxx.add(('1_samuel', 18, 11))
for v in range(17, 19+1):
    hb_no_lxx.add(('1_samuel', 18, v))
hb_no_lxx.add(('1_samuel', 18, 30))

# 1 Sam 13:1 - absent in LXX Sinaiticus (present in Lucianic)
# Depending on your LXX source this may or may not be present
# hb_no_lxx.add(('1_samuel', 13, 1))  # uncomment if your LXX source lacks it

print(f"1 Samuel: {len([k for k in hb_to_lxx if k[0]=='1_samuel'])} verse remaps, "
      f"{len([k for k in hb_no_lxx if k[0]=='1_samuel'])} MT verses absent in LXX")
````

````python
def resolve_lxx_key(hb_key):
    """
    Given a Hebrew MT verse key (book, chapter, verse),
    return the corresponding LXX verse key for Greek text lookup.
    
    Returns:
        tuple: (lxx_key, status)
        status: 'identity' | 'remapped' | 'no_lxx' | 'merged'
    """
    if hb_key in hb_no_lxx:
        return None, 'no_lxx'
    if hb_key in hb_to_lxx:
        return hb_to_lxx[hb_key], 'remapped'
    # Identity: Hebrew key == LXX key
    return hb_key, 'identity'

# Test with the problematic example from the user
test_cases = [
    ('1_samuel', 21, 1),   # Hebrew 21:1 -> should map to LXX 20:42
    ('1_samuel', 21, 2),   # Hebrew 21:2 -> should map to LXX 21:1
    ('1_samuel', 24, 1),   # Hebrew 24:1 -> should map to LXX 23:29
    ('1_samuel', 17, 12),  # Should be absent in LXX
    ('1_samuel', 1, 1),    # Identity mapping
]
for tc in test_cases:
    lxx_key, status = resolve_lxx_key(tc)
    print(f"  Heb {tc} -> LXX {lxx_key} [{status}]")
````

## Integration: update Greek text lookup to use hb_to_lxx mapping

The main save loop (section 3) currently does:
```python
if key in strongs_g.groups.keys():
    greek_text = " ".join(strongs_g.get_group(key)...)
```

This assumes Hebrew key == Greek key, which fails when versification differs.
We patch it to use `resolve_lxx_key()` first.

````python
# ============================================================
# Merge support for Hebrew->LXX where one Hebrew verse maps to
# part of an LXX verse (e.g. Hebrew 21:1 -> tail of LXX 20:42)
# ============================================================
# When hb_to_lxx maps a Hebrew verse to an LXX verse that ALSO
# has its own Hebrew source (e.g. Heb 21:1 -> LXX 20:42, but
# Heb 20:42 also -> LXX 20:42), we have a merge situation.
# The Greek text at LXX 20:42 covers both Hebrew 20:42 + 21:1.
#
# hb_to_lxx_merge tracks these: {lxx_key: [hb_key1, hb_key2, ...]}

from collections import defaultdict
hb_to_lxx_merge = defaultdict(list)

# Build merge map: find all LXX keys that have multiple Hebrew sources
lxx_target_sources = defaultdict(list)
# First, all identity mappings (Hebrew key == LXX key, unless overridden)
# We only care about keys that appear as targets of explicit remaps
for hb_key, lxx_key in hb_to_lxx.items():
    lxx_target_sources[lxx_key].append(hb_key)

# Check which LXX targets also have an identity-mapped Hebrew source
for lxx_key, hb_sources in lxx_target_sources.items():
    # Does lxx_key also serve as an identity key for some Hebrew verse?
    # i.e., is there a Hebrew verse == lxx_key that is NOT in hb_to_lxx?
    if lxx_key not in hb_to_lxx:  # lxx_key is not itself remapped elsewhere
        # Then lxx_key is both: identity for itself AND target of a remap
        hb_to_lxx_merge[lxx_key] = [lxx_key] + hb_sources  # identity first, then remapped

if hb_to_lxx_merge:
    print(f"LXX merge situations (1 LXX verse <- multiple Hebrew verses): {len(hb_to_lxx_merge)}")
    for lxx_k, hb_sources in sorted(hb_to_lxx_merge.items())[:10]:
        print(f"  LXX {lxx_k} <- Hebrew {hb_sources}")
else:
    print("No LXX merge situations detected.")
````

````python
def get_greek_text_for_hb_key(hb_key, strongs_g):
    """
    Given a Hebrew MT verse key, look up the corresponding Greek (LXX) text.
    
    Handles:
    1. Identity mapping (Hebrew key == LXX key)
    2. Verse remapping (different chapter/verse in LXX)
    3. Verses absent in LXX (returns '-')
    4. Merge situations (one LXX verse covers multiple Hebrew verses)
    
    Returns:
        str: Greek text, or '-' if not available
    """
    lxx_key, status = resolve_lxx_key(hb_key)
    
    if status == 'no_lxx':
        return '-'
    
    # Check if this LXX key is a merge target
    # If so, the entire LXX verse text belongs to this Hebrew verse
    # (or we'd need to split it -- for now we return the full LXX verse)
    if lxx_key in strongs_g.groups:
        return " ".join(
            strongs_g.get_group(lxx_key).sort_values("word")["form"].astype(str)
        )
    
    # If the LXX key was remapped but no Greek text exists at that key either,
    # try the original Hebrew key as fallback (for Apostolic Bible / BibleHub sources
    # that may use Hebrew numbering for Greek text)
    if status == 'remapped' and hb_key in strongs_g.groups:
        return " ".join(
            strongs_g.get_group(hb_key).sort_values("word")["form"].astype(str)
        )
    
    return '-'

print("get_greek_text_for_hb_key() defined.")
print("This function will be used in section 3 save loop instead of direct strongs_g lookup.")
````
