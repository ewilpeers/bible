```python
def diff_hebrew_vs_json(ref_hb_words, mapping_hb_words, key):
    aminlen = min(len(ref_hb_words), len(mapping_hb_words))
    
    print(f"{RED_W}⚠ Hebrew mismatch at {key}{RESET}")
    print(f"  ref words: {len(ref_hb_words)}  json words: {len(mapping_hb_words)}")
    print(f"  {'#':<4} {'ref':<30} {'json'}")

    for j in range(aminlen):
        ref_str  = str(ref_hb_words.iloc[j]["form"])
        json_str = mapping_hb_words[j].get("hebrew", "")
        inum= int(mapping_hb_words[j].get("index", str(j+1)))

        if ref_str == json_str:
            # identical — print plain
            print(f"  {j+1:<4} {GREEN}{ref_str}{RESET}")
        else:
            # diff char by char
            minlen = min(len(ref_str), len(json_str))
            s_ref = s_json = ""
            for i in range(minlen):
                if ref_str[i] == json_str[i]:
                    s_ref  += f"{GREEN}{ref_str[i]}{RESET}"
                    s_json += f"{GREEN}{json_str[i]}{RESET}"
                else:
                    s_ref  += f"{DIFF_RED}{ref_str[i]}{RESET}"
                    s_json += f"{DIFF_RED}{json_str[i]}{RESET}"
            if len(ref_str) > minlen:
                s_ref  += f"{DIFF_BLUE}{ref_str[minlen:]}{RESET}"
            if len(json_str) > minlen:
                s_json += f"{DIFF_BLUE}{json_str[minlen:]}{RESET}"
            # one word per line to avoid bidi chaos
            print(f"  {inum:<4} ref : {s_ref}")
            print(f"  {'':4} json: {s_json}")

    # surplus words
    if len(ref_hb_words) > aminlen:
        for j in range(aminlen, len(ref_hb_words)):
            form = str(ref_hb_words.iloc[j]["form"])
            print(f"  {j+1:<4} {DIFF_BLUE}ref only : {form}{RESET}")
    if len(mapping_hb_words) > aminlen:
        for j in range(aminlen, len(mapping_hb_words)):
            heb = mapping_hb_words[j].get("hebrew", "")
            print(f"  {j+1:<4} {DIFF_BLUE}json only: {heb}{RESET}")
```

```python
def split_str_tokened(input_string, tokens):
    result = []
    current_string = input_string
    
    for token in tokens:
        # Atrodam pirmo marķiera parādīšanās vietu
        token_index = current_string.find(token)
        #print(token_index)
        #print(f"::str:: {current_string} \n::token::  {token}")
        # Ja marķieris netiek atrasts, pārtraucam procesu
        if token_index == -1:
            break
            
        # Sadalām virkni: 
        # pirmā daļa līdz marķiera beigām (ieskaitot pašu marķieri)
        # otrā daļa - viss, kas paliek aiz marķiera
        split_pos = token_index + len(token)
        part = current_string[:split_pos]
        current_string = current_string[split_pos:].strip()

        result.append(part)
        

    if current_string:
        result.append(current_string)
    #print(result)
    #print("-"*50)
    #print(tokens)
    return result
```

```python
import pandas as pd
import time
from pathlib import Path
import json
import os
import json
import re
from collections import Counter
import unicodedata

DRY_RUN=1

def validate_latvian_usage(verse_text, mappings, leftover_latvian_from_verse_data):
    """
    Pārbauda, vai visi vārdi no latviešu teksta ir atrodami kartējumos.
    Izmanto specifisku latviešu burtu kopu, lai novērstu vārdu sašķelšanos.
    """
    # 1. Normalizējam tekstu uz NFC (lai 'ī' būtu viens simbols, nevis i + macron)
    ####print(verse_text)
    verse_text = unicodedata.normalize('NFC', verse_text)
    
    # Definējam latviešu burtu kopu (lielie un mazie)
    lv_chars = "A-Za-zāēīūģķļņšžčĀĒĪŪĢĶĻŅŠŽČ"
    lv_regex = f"[{lv_chars}]+"

    # 2. Tokenizējam tekstu
    text_words = re.findall(lv_regex, verse_text)
    
    # 3. Savācam visus vārdus no kartējumiem
    mapped_words = []
    
    for m in mappings:
        # Pārbaudām galvenos kartējumus
        if m.get('latvian'):
            for item in m['latvian']:
                if item and item != "-":
                    #TODO: ja ir neatzīst un atzīst orig tulkojumos vienlaikus...
                    # Arī šeit normalizējam un meklējam ar lv_regex
                    clean_item = unicodedata.normalize('NFC', item)
                    #print(re.findall(lv_regex, clean_item))
                    mapped_words.extend(re.findall(lv_regex, clean_item))
        
        if leftover_latvian_from_verse_data:
            for item in leftover_latvian_from_verse_data:
                clean_item = unicodedata.normalize('NFC', item)
                mapped_words.extend(re.findall(lv_regex, clean_item))
    
    text_counts = Counter(text_words)
    mapped_counts = Counter(mapped_words)

    missing = []
    for word, count in text_counts.items():
        if mapped_counts[word] < count:
            lw = word.lower()
            
            # Loģika priedēklim "ne-" (piemēram, 'atzīst' vs 'neatzīst')
            if len(lw) > 2 and lw.startswith("ne"):
                stem = word[2:]  # Saglabājam oriģinālo reģistru saknei
                if mapped_counts[stem] >= count:
                    continue
            
            # Pievienojam trūkstošos gadījumus
            for _ in range(count - mapped_counts[word]):
                missing.append(word)
            
    return missing

def split_chapter_json(book, chapter_num, hb_g, strongs_g, lv_g):
    # Path to the source chapter file: bible/genesis/1.json
    source_path = Path("bible") / book.lower() / f"{chapter_num}.json"
    
    if not source_path.exists():
        print(f"❓ Source {source_path} not found. Skipping.")
        return

    with open(source_path, "r", encoding="utf-8") as f:
        verses_list = json.load(f)
    
    print(verses_list[0])
    dest_dir = Path("bible") / book.lower() / str(chapter_num)
    if not DRY_RUN:
        dest_dir.mkdir(parents=True, exist_ok=True)
    
    for i, verse_data in enumerate(verses_list):
        verse_number = int(i+1)
        print(verse_number)
        print("*"*80)
        key = (book, int(chapter_num), verse_number)

        # 1. Get Reference Data
        if key not in hb_g.groups:
            print(f"⚠️ Missing Hebrew reference data for {key}.")
            raise Exception(f"HBRW missing - no such chapter {key}")
        else:
            ref_hb = hb_g.get_group(key).sort_values("word")
            
        if key not in strongs_g.groups:
            print(f"⚠️ Missing STRONGS reference data for {key}.")
            ref_strongs=[]
        else:
            ref_strongs = strongs_g.get_group(key).sort_values("word")

        l65_text=""
        if key in hb_to_l65.keys():
            kkey = hb_to_l65[key]
            if(kkey == None):
                l65_text="-"
            else:
                if not kkey[0][0]=="!":
                    l65_text=" ".join(l65_g.get_group(kkey).sort_values("word_idx")["text"].astype(str))
                else:
                    nkey = (kkey[0][1:], kkey[1], kkey[2])
                    #print("here")
                    #print(nkey)
                    for k, v in l65_to_hb_merge.items():
                        #merge will ALWAY have at least 2 items in key (either mergable verse Nr2 or splitstring Nr1)
                        if(isinstance(k[1], str) and k[0] == nkey):
                            
                            #case 1: lv verse has multiple hebrw verses in it
                            sstr = " ".join(l65_g.get_group(nkey).sort_values("word_idx")["text"].astype(str))
                            spArr = split_str_tokened(sstr, [k[i] for i in range(1, len(k))])
                            #print(spArr)
                            if(len(spArr) != len(v)):
                                raise Exception (f"could not tokenise properly!!!@ {k} to {v}")
                            for i, vsNm in enumerate(v):
                                if vsNm == key:
                                    l65_text=spArr[i]
                                    break
                            if not l65_text:
                                raise Exception (f"could not find verse part for the {key} at {k} to {v}")
                        elif(v==key):
                            #now this is inverse case of the above, l65 verses should be merged.
                            rslt = []
                            for mrg_verse in k:
                                rslt.append(" ".join(l65_g.get_group(mrg_verse).sort_values("word_idx")["text"].astype(str)))
                            l65_text = " ".join(rslt)
                            break
                    if not  l65_text:
                        raise Exception (f"key not found! {nkey} :: {key}")
        else:
            l65_text=" ".join(l65_g.get_group(key).sort_values("word_idx")["text"].astype(str))
        

        # 2. Validation: Hebrew Word Count and Order
        mapping_hb_words = verse_data.get('hebrew_words', [])
        if len(ref_hb) != len(mapping_hb_words):
            print(f"❌ Hebrew word count mismatch at {key}: Ref {len(ref_hb)} vs JSON {len(mapping_hb_words)}")
            diff_hebrew_vs_json(ref_hb, mapping_hb_words, key)
            #print(mapping_hb_words)
            #continue
        leftover_latvian_from_verse_data = verse_data.get('leftover_latvian', [])
        # 3. Validation: Latvian Usage
        latvian_missing = validate_latvian_usage(l65_text, mapping_hb_words, leftover_latvian_from_verse_data)
        if latvian_missing:
            print(f"⚠️ Latvian mapping incomplete at {key}: Missing {', '.join(latvian_missing)}")
            # Optional: continue/break based on how strict you want to be

        # 4. Construct the individual verse object
        # Note: We include the 'locus' as your build_verse_object expects it
        leftover_latvian = verse_data.get("leftover_latvian", [])
        leftover_latvian.extend(latvian_missing)
        verse_output = {
            "locus": f"{book}_{chapter_num}_{i+1}",
            "hebrew_words": mapping_hb_words,
            "leftover_latvian": leftover_latvian
        }

        if not DRY_RUN:
        # 5. Write to File: bible/luke/1/luke_1_1.txt
            print(i+1)
            file_name = f"{book}_{chapter_num}_{i+1}.txt"
            with open(dest_dir / file_name, "w", encoding="utf-8") as out_f:
                json.dump(verse_output, out_f, ensure_ascii=False, indent=4)

    print(f"✅ Successfully split {source_path} into {dest_dir}")
```
