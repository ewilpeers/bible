```python
def split_str_tokened(input_string, tokens):
    result = []
    current_string = input_string
    
    for token in tokens:
        # Atrodam pirmo marķiera parādīšanās vietu
        token_index = current_string.find(token)
        #print(token_index)
        #print(f"::str:: {current_string} \n::token::  {token}")
        # Ja marķieris netiek atrasts, pārtraucam procesu
        if token_index == -1:
            break
            
        # Sadalām virkni: 
        # pirmā daļa līdz marķiera beigām (ieskaitot pašu marķieri)
        # otrā daļa - viss, kas paliek aiz marķiera
        split_pos = token_index + len(token)
        part = current_string[:split_pos]
        current_string = current_string[split_pos:].strip()

        result.append(part)
        

    if current_string:
        result.append(current_string)
    #print(result)
    #print("-"*50)
    #print(tokens)
    return result
```

```python
def lv_get_key_fulltext(key, sett):
    l65_text=""
    if key in hb_to_l65.keys():
        kkey = hb_to_l65[key]
        if(kkey == None):
            l65_text="-"
        else:
            if not kkey[0][0]=="!":
                l65_text=" ".join(sett.get_group(kkey).sort_values("word_idx")["form"].astype(str))
            else:
                nkey = (kkey[0][1:], kkey[1], kkey[2])
                #print("here")
                #print(nkey)
                for k, v in l65_to_hb_merge.items():
                    #merge will ALWAY have at least 2 items in key (either mergable verse Nr2 or splitstring Nr1)
                    if(isinstance(k[1], str) and k[0] == nkey):
                        
                        #case 1: lv verse has multiple hebrw verses in it
                        sstr = " ".join(sett.get_group(nkey).sort_values("word_idx")["form"].astype(str))
                        spArr = split_str_tokened(sstr, [k[i] for i in range(1, len(k))])
                        #print(spArr)
                        if(len(spArr) != len(v)):
                            raise Exception (f"could not tokenise properly!!!@ {k} to {v}")
                        for i, vsNm in enumerate(v):
                            if vsNm == key:
                                l65_text=spArr[i]
                                break
                        if not l65_text:
                            raise Exception (f"could not find verse part for the {key} at {k} to {v}")
                    elif(v==key):
                        #now this is inverse case of the above, l65 verses should be merged.
                        rslt = []
                        for mrg_verse in k:
                            rslt.append(" ".join(sett.get_group(mrg_verse).sort_values("word_idx")["form"].astype(str)))
                        l65_text = " ".join(rslt)
                        break
                if not  l65_text:
                    raise Exception (f"key not found! {nkey} :: {key}")
    else:
        if key in sett.groups:
            l65_text=" ".join(sett.get_group(key).sort_values("word_idx")["form"].astype(str))
        else:
            l65_text="-"
    return l65_text
```

```python
POS_MAP = {
    "V": "Verb",
    "N": "Noun",
    "Adv": "Adverb",
    "Adj": "Adjective",
    "Art": "Article",
    "DPro": "Demonstrative Pronoun",
    "IPro": "Interrogative / Indefinite Pronoun",
    "PPro": "Personal / Possessive Pronoun",
    "RecPro": "Reciprocal Pronoun",
    "RelPro": "Relative Pronoun",
    "RefPro": "Reflexive Pronoun",
    "Prep": "Preposition",
    "Conj": "Conjunction",
    "I": "Interjection",
    "Prtcl": "Particle",
    "PrtclDsj": "Particle, Disjunctive Particle",
    "Heb": "Hebrew Word",
    "Aram": "Aramaic Word"
}

PERSON_MAP = {"1": "1st Person", "2": "2nd Person", "3": "3rd Person"}
TENSE_MAP = {"P": "Present", "I": "Imperfect", "F": "Future",
             "A": "Aorist", "R": "Perfect", "L": "Pluperfect"}
MOOD_MAP = {"I": "Indicative", "M": "Imperative",
            "S": "Subjunctive", "O": "Optative",
            "N": "Infinitive", "P": "Participle"}
VOICE_MAP = {"A": "Active", "M": "Middle", "P": "Passive", "M/P": "Middle or Passive"}
CASE_MAP = {"N": "Nominative", "V": "Vocative", "A": "Accusative",
            "G": "Genitive", "D": "Dative"}
NUMBER_MAP = {"S": "Singular", "P": "Plural"}
GENDER_MAP = {"M": "Masculine", "F": "Feminine", "N": "Neuter"}
COMPARISON_MAP = {"C": "Comparative", "S": "Superlative"}
```

```python
def parse_morph_code(code):
    if not isinstance(code, str) or not code:
        return {}

    parts = code.split("-")
    pos_key = parts[0]
    
    result = {"part_of_speech": POS_MAP.get(pos_key, pos_key)}

    if len(parts) < 2:
        return result

    # --- VERB LOGIC ---
    if pos_key == "V":
        tmv = parts[1]
        if len(tmv) >= 3:
            result["tense"] = TENSE_MAP.get(tmv[0])
            result["mood"] = MOOD_MAP.get(tmv[1])
            result["voice"] = VOICE_MAP.get(tmv[2])
        
        if len(parts) >= 3:
            third_part = parts[2]
            # FIX: Participles have Case-Gender-Number, not Person-Number
            if result.get("mood") == "Participle":
                for char in third_part:
                    if char in CASE_MAP:
                        result["case"] = CASE_MAP.get(char)
                    elif char in GENDER_MAP:
                        result["gender"] = GENDER_MAP.get(char)
                    elif char in NUMBER_MAP:
                        result["number"] = NUMBER_MAP.get(char)
            else:
                # Finite verbs: Person-Number
                if len(third_part) >= 2:
                    result["person"] = PERSON_MAP.get(third_part[0])
                    result["number"] = NUMBER_MAP.get(third_part[1])
        return result

    # --- NOUN / ADJ / ART / PRONOUN LOGIC ---
    details = parts[1]
    for char in details:
        if char in CASE_MAP:
            result["case"] = CASE_MAP.get(char)
        elif char in GENDER_MAP:
            result["gender"] = GENDER_MAP.get(char)
        elif char in NUMBER_MAP:
            result["number"] = NUMBER_MAP.get(char)
        elif char in COMPARISON_MAP:
            result["comparison"] = COMPARISON_MAP.get(char)
        elif char in PERSON_MAP and "Pro" in pos_key:
            result["person"] = PERSON_MAP.get(char)

    return result
```

```python
def parse_hebrew_morph_code(code):
    """
    Parse a BibleHub Hebrew morphology tag like:
      Conj‑w|V‑Qal‑ConsecImperf‑3ms
      N‑proper‑ms
      Pro|V‑Qal‑Inf
    Returns a list of dicts, one per pipe-separated segment.
    """
    if not isinstance(code, str) or not code:
        return []

    STEMS = {
        "Qal": "Qal", "Nifal": "Nifal", "Piel": "Piel", "Pual": "Pual",
        "Hiphil": "Hiphil", "Hophal": "Hophal", "Hithpael": "Hithpael",
        "QalPass": "Qal Passive",
    }
    ASPECTS = {
        "Perf": "Perfect", "Imperf": "Imperfect",
        "ConjPerf": "Conjunctive Perfect", "ConjImperf": "Conjunctive Imperfect",
        "ConsecImperf": "Consecutive Imperfect",
        "Imp": "Imperative", "Prtcpl": "Participle",
        "QalPassPrtcpl": "Qal Passive Participle",
        "Inf": "Infinitive Construct", "InfAbs": "Infinitive Absolute",
    }
    PERSON = {"1": "1st", "2": "2nd", "3": "3rd"}
    GENDER = {"m": "masculine", "f": "feminine", "c": "common"}
    NUMBER = {"s": "singular", "p": "plural", "d": "dual"}
    STATE  = {"c": "construct", "d": "determinate"}
    CONJ_TYPES = {
        "w": "waw", "b": "be (in)", "k": "ke (according to)",
        "l": "le (with regard to)", "m": "min (from)",
    }

    def parse_pgn(pgn):
        """Parse person-gender-number string like '3ms', '2fp', '1cs'."""
        result = {}
        i = 0
        # person
        if i < len(pgn) and pgn[i] in PERSON:
            result["person"] = PERSON[pgn[i]]
            i += 1
        # gender
        if i < len(pgn) and pgn[i] in GENDER:
            result["gender"] = GENDER[pgn[i]]
            i += 1
        # number
        if i < len(pgn) and pgn[i] in NUMBER:
            result["number"] = NUMBER[pgn[i]]
            i += 1
        # paragogic nun suffix
        if pgn[i:] == "Pn":
            result["suffix"] = "Paragogic nun"
        return result

    def parse_gn(gn):
        """Parse gender-number string like 'ms', 'fp', 'cs'."""
        result = {}
        i = 0
        if i < len(gn) and gn[i] in GENDER:
            result["gender"] = GENDER[gn[i]]
            i += 1
        if i < len(gn) and gn[i] in NUMBER:
            result["number"] = NUMBER[gn[i]]
            i += 1
        if gn[i:] in STATE:
            result["state"] = STATE[gn[i:]]
        return result

    def parse_segment(seg):
        parts = seg.split("‑")  # non-breaking hyphen used by BibleHub
        if not parts:
            return {"raw": seg}

        pos = parts[0]
        result = {}
        # --- Pronominal suffix (standalone segment like "3fs", "1cs") ---
        if len(parts) == 1 and parts[0][0] in PERSON:
            result["pos"] = ""
            result.update(parse_pgn(parts[0]))
            return result
        # --- Conjunction ---
        if pos == "Conj":
            result["pos"] = "Conjunction"
            if len(parts) > 1 and parts[1] in CONJ_TYPES:
                result["type"] = CONJ_TYPES[parts[1]]
            return result

        # --- Preposition ---
        if pos == "Prep":
            result["pos"] = "Preposition"
            if len(parts) > 1 and parts[1] in CONJ_TYPES:
                result["type"] = CONJ_TYPES[parts[1]]
            return result

        # --- Article ---
        if pos == "Art":
            result["pos"] = "Definite Article"
            return result

        # --- Direct Object Marker ---
        if pos == "DirObjM":
            result["pos"] = "Direct Object Marker"
            return result

        # --- Interjection ---
        if pos == "I":
            result["pos"] = "Interjection"
            return result

        # --- Interrogative ---
        if pos == "Interrog":
            result["pos"] = "Interrogative"
            return result

        # --- Adverb ---
        if pos == "Adv":
            result["pos"] = "Adverb"
            if len(parts) > 1 and parts[1] == "NegPrt":
                result["type"] = "Negative Particle"
            return result

        # --- Pronoun ---
        if pos == "Pro":
            result["pos"] = "Pronoun"
            if len(parts) > 1:
                if parts[1] == "r":
                    result["type"] = "Relative"
                else:
                    result.update(parse_pgn(parts[1]))
                    if len(parts) > 2 and parts[2] == "Pn":
                        result["suffix"] = "Paragogic nun"
            return result

        # --- Noun / Adjective / Number ---
        if pos in ("N", "Adj", "Number"):
            POS_LABEL = {"N": "Noun", "Adj": "Adjective", "Number": "Number"}
            result["pos"] = POS_LABEL[pos]
            if len(parts) < 2:
                return result
            subtype = parts[1]
            if subtype == "proper":
                result["type"] = "Proper Noun"
                if len(parts) > 2:
                    result.update(parse_gn(parts[2]))
            elif subtype == "g":
                result["type"] = "Gentilic"
                if len(parts) > 2:
                    result.update(parse_gn(parts[2]))
                if len(parts) > 3 and parts[3] in STATE:
                    result["state"] = STATE[parts[3]]
            elif subtype == "o":
                result["type"] = "Ordinal"
                if len(parts) > 2:
                    result.update(parse_gn(parts[2]))
                if len(parts) > 3 and parts[3] in STATE:
                    result["state"] = STATE[parts[3]]
            else:
                # regular noun: parts[1] is gender-number[-state]
                result.update(parse_gn(subtype))
                if len(parts) > 2 and parts[2] in STATE:
                    result["state"] = STATE[parts[2]]
            return result

        # --- Verb ---
        if pos == "V":
            result["pos"] = "Verb"
            if len(parts) < 2:
                return result
            stem = parts[1]
            result["stem"] = STEMS.get(stem, stem)
            if len(parts) < 3:
                return result
            aspect = parts[2]
            # handle cohortative / jussive suffixes
            for suffix in (".Cohort", ".h", ".Jus"):
                if aspect.endswith(suffix):
                    result["modifier"] = {
                        ".Cohort": "Cohortative",
                        ".h": "Cohortative (assumed)",
                        ".Jus": "Jussive",
                    }[suffix]
                    aspect = aspect[: -len(suffix)]
                    break
            result["aspect"] = ASPECTS.get(aspect, aspect)
            if len(parts) < 4:
                return result
            pgn = parts[3]
            # Imperative / Participle use gender-number only
            if result["aspect"] in ("Imperative", "Participle", "Qal Passive Participle"):
                result.update(parse_gn(pgn))
            else:
                result.update(parse_pgn(pgn))
            return result

        # fallback
        return {"raw": seg}

    # Split on pipe — each segment is an independent morphological unit
    segments = code.split("|")
    parsed = [parse_segment(s.strip()) for s in segments]
    return parsed if len(parsed) > 1 else parsed[0]


def format_hebrew_morph(code):
    """Human-readable one-liner from a Hebrew morph code."""
    parsed = parse_hebrew_morph_code(code)
    if isinstance(parsed, dict):
        parsed = [parsed]
    parts = []
    for p in parsed:
        if "raw" in p:
            parts.append(p["raw"])
            continue
        tokens = []
        for k in ("pos", "type", "stem", "aspect", "modifier", "person", "gender", "number", "state", "suffix"):
            if k in p:
                if k == "type" and p["type"] == "Proper Noun":
                    tokens.append("Proper")
                else:
                    tokens.append(p[k])
        parts.append(" ".join(tokens))
    return " | ".join(parts)
```
